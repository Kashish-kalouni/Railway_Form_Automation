```python
from rich import print
import csv
from rich.console import Console
from rich.panel  import Panel
from fpdf import FPDF
import textwrap


class Registration:
    def __init__(self):
        self.name = input("Enter name's (separated by ',' ) : ").split(",")
        
        self.n =[i.upper() for i in self.name]
        
        self.age_validation()
        
        self.sex_validation()
        
        self.child = input("Enter children name's (separated by ',' ) : ").split(",")
        
        self.child_name =[i.upper() for i in self.child]
        
        if self.child == ['']:
            self.child = []
            self.child_name = []
        else:
            self.child_age_validation()
            self.child_sex_validation() 
            
        
        
        self.concession = input("Enter consession's (separated by ',') :").split(",")
        
        self.medical = input("If you are a Medical Practitioner, please write (Yes or No) : ")
        
        self.extra = input("Do you want to be upgraded without any extra charge?   Write Yes/No : ")
        
        self.train_no = input("Train No & Name : ")
        
        self.date = input("Date of Journey (dd/mm/yyyy) : ")
        
        self.classno = input("Class : ")
        
        self.seat = input("No. of Berth / Seat : ")
        
        self.seat1 = input("Enter berth No. (separated by ',') : ").split(",")
        
        self.from1 = input("Station From : ")
        
        self.to1 = input("Station To : ")
        
        self.boarding = input("Boarding At : ")
        
        self.rev_upto = input("Reservation Upto : ")
        
        self.address = input("Full address : ")

        
        while True:
            self.mobile = input("Telephone no: ")
            if self.mobile.isdigit() and len(self.mobile) == 10:
                break
            else:
                print("Invalid mobile number! Please enter a 10-digit number.")
            
        self.time = input("Time : ")
        
        self.choice_b = input("Enter all Choice seperated by ',' (Lower/Upper Berth) : ").split(",")
        
        self.choice_f = input("Choice (Veg/Non-veg, etc.) : ")
        
        self.requisition_no = input("Enter Requisition No. : ")
        
        self.pnr_no = input("Enter PNR No. : ")
        
        self.amount = int(input("Enter the amount for one ticket : "))

    def age_validation(self):
        while True:
            self.age = input("Enter ages (separated by ',' ): ").split(",")
            try:
                self.age = [int(i) for i in self.age]
                if len(self.age) != len(self.n):
                    print("Number of ages must match number of names!")
                    continue
                if min(self.age) > 5:
                    break
                else:
                    print("All ages must be greater than 5.")
            except ValueError:
                print("Please enter valid numbers for age.")
    
    def sex_validation(self):
        while True:
            self.sex = input("Enter genders (separated by ',' M/F): ").split(",")
            self.sex1 = [i.upper() for i in self.sex]
            if len(self.sex1) != len(self.n):
                print("Number of genders must match number of names!")
                continue
            if set(self.sex1).issubset({"M", "F"}):
                break
            else:
                print("Invalid input! Please enter only 'M' or 'F'.")
    
    def child_age_validation(self):
        while True:
            self.age_c = input("Enter children's ages (separated by ',' ):").split(",")
            try:
                self.age_c = [int(i) for i in self.age_c]
                if len(self.age_c) != len(self.child_name):
                    print("Number of children's ages must match number of children's names!")
                    continue
                if max(self.age_c) < 5:
                    break
                else:
                    print("All children's ages must be less than 5.")
            except ValueError:
                print("Please enter valid numbers for age.")
    
    def child_sex_validation(self):
        while True:
            self.sex_child = input("Enter children's genders (separated by ',' M/F):").split(",")
            self.sex_child = [i.upper() for i in self.sex_child]
            if len(self.sex_child) != len(self.child_name):
                print("Number of children's genders must match number of children's names!")
                continue
            if set(self.sex_child).issubset({"M", "F"}):
                break
            else:
                print("Invalid input! Please enter only 'M' or 'F'.")

    def show(self):
        first = "[bold] RAILWAY REGISTRATION / CANCELLATION REQUISTION FORM [/bold]\n"
        output = f"""{first: ^120} 

If you are a Medical Practitioner. Please write(Yes or No) Dr. :[bold] {self.medical} [/bold]   (You could be of help in an emergency)

If you want Sr. Citizen concession, please write (Yes/No):[bold] {self.concession[0]} [/bold]  (if yes, please carry a proof of age during the journey to avoid inconvenience of penal charging under extant Railway Rules)

Do you want to be upgraded without any extra charge? Write (Yes/No):[bold] {self.extra} [/bold] (Ifthis option is not exercised, full fare paying passengers may be upgraded automatically)


Train No & Name:[bold] {self.train_no} [/bold] \t Date of Journey:[bold] {self.date} [/bold]
Class:[bold] {self.classno} [/bold] \t\t\t\t No. of Berth / Seat:[bold] {self.seat} [/bold]
Station From:[bold] {self.from1} [/bold] \t\t\t\t Station To:[bold] {self.to1} [/bold]
Boarding At:[bold] {self.boarding} [/bold] \t\t\t Reservation Upto:[bold] {self.rev_upto} [/bold]

                                                                                                                             
[bold]S.No.   Name in Block Letters    Sex    Age    Concession    Choice if any [/bold]

"""
        for i in range(len(self.n)) : 
            output += f"{i+1} \t {self.n[i]} \t\t\t {self.sex[i]} \t {self.age[i]}\t{self.concession[i]}\t\t{self.choice_b[i]}\n" 


        output += "\n\n[bold]CHILDREN BELOW 5 YEARS(FOR WHOM TICKET TO BE NOT ISSUED)[/bold]\n\n[bold]S.No. \t Name in Block Letters \tSex \t Age[/bold]\n\n"
        for i in range(len(self.child_name)) : 
            output += f"{i+1} \t {self.child_name[i]} \t\t\t {self.sex_child[i]} \t {self.age_c[i]}\n"

        sig_app = "[bold]Signature of Applicant: ___________________________[/bold]\n"
        second  = "[bold]ONWARD / RETURN JOURNEY DETAILS [/bold]"
        self.f_name = " , ".join(self.n)
        self.f_seat = " , ".join(self.seat1)
        output += "\n\n"
        output += f"""{second : ^120}
        
        
Train No & Name : {self.train_no} \t\t Date : {self.date}
Class : {self.classno}        \t\t\t\t Station From : {self.from1}  \t\t\tTo : {self.to1}

Name of Applicant : {self.f_name}
Full Address      : {self.address}
Telephone No      : {self.mobile} \t\t Date : {self.date} \t\t Time : {self.time}

{sig_app:>120}
"""

        sig_res = "[bold]Signature of Reservation Clerk: ___________________________[/bold]\n\n\n"
        total_amount = len(self.name) * self.amount
        third = "[bold]FOR OFFICE USE ONLY[/bold]\n\n"
        output += f"""{third : ^120}
Requisition No. : {self.requisition_no} \t\t\t\t PNR No. : {self.pnr_no}
Berth/Seat No.  : {self.f_seat} \t\t  Amount Collected: {total_amount}

[bold]{sig_res:>120}

Notes:
1. Maximum permissible passengers are 6 per requisition.
2. One person can give one requisition form at a time.
3. Please check your ticket and balance amount before leaving the window.
4. Forms not properly filled or in illegible forms shall not be entertained.
5. Choice is subject to availability.
[/bold]
"""
        self.final_output = output
        return output
        

    def save_to_csv(self):
        adult_data = [[i+1, self.n[i], self.sex[i], self.age[i], self.concession[i], self.choice_b[i]]
                      for i in range(len(self.n))]
        child_data = [[i+1, self.child_name[i], self.sex_child[i], self.age_c[i]]
                      for i in range(len(self.child_name))]

        filename = "registration_details.csv"
        with open(filename, mode="w", newline="") as file:
            writer = csv.writer(file)

            writer.writerow(["Adult Data"])
            writer.writerow(["S.No.", "Name", "Sex", "Age", "Concession", "Choice"])
            writer.writerows(adult_data)
            writer.writerow([])

            if child_data:
                writer.writerow(["Child Data"])
                writer.writerow(["S.No.", "Name", "Sex", "Age"])
                writer.writerows(child_data)

        print(f"\nData saved successfully in '{filename}'!")

    def save_to_pdf(self):
        pdf = FPDF()
        pdf.add_page()

        pdf.set_font("Helvetica", "B", 16)
        pdf.cell(0, 10, "Railway Registration / Cancellation Form",
                 new_x="LMARGIN", new_y="NEXT", align='C')
        pdf.ln(10)
        pdf.set_font("Helvetica", size=10)

        form_text = self.final_output.replace("[bold]", "").replace("[/bold]", "")

        lines = form_text.split("\n")

        lines = lines[1:]

        for raw_line in lines:
            if not raw_line.strip():
                pdf.ln(6)
                continue

            if "Signature of" in raw_line:
                pdf.cell(0, 8, raw_line.strip(),
                         new_x="LMARGIN", new_y="NEXT")
                continue

            for chunk in textwrap.wrap(raw_line.rstrip(), width=100):
                pdf.multi_cell(0, 8,
                               text=chunk,
                               new_x="LMARGIN", new_y="NEXT")

        pdf.output("registration_form.pdf")
        print("\nPDF generated successfully as 'registration_form.pdf'!")



import csv
from fpdf import FPDF

class Cancellation:
    def __init__(self):
        self.names = input("Enter your Name's (separated by ',' ): ").split(",")
        self.names = [i.strip().upper() for i in self.names]
        self.pnr_no = input("Enter PNR No: ")
        self.journey_date = input("Enter Journey Date (dd/mm/yyyy): ")
        self.ticket_amount = int(input("Enter Ticket Amount: "))
        self.reason = input("Reason for Cancellation: ")
        self.refund_mode = input("Refund Mode (Bank Transfer / Wallet): ")

    def show(self):
        first = "[bold]Railway Ticket Cancellation Form[/bold]"
        second = "Signature of Passenger(s): ___________________"
        
        names_formatted = ", ".join(self.names)

        self.final_output = f"""
        {first:^120}
        
        Name(s): {names_formatted}
        
        PNR No: {self.pnr_no}
        
        Journey Date: {self.journey_date}
        
        Ticket Amount: {self.ticket_amount} Rs
        
        Reason for Cancellation: {self.reason}
        
        Refund Mode: {self.refund_mode}



        {second:>100}
        """
        return self.final_output

    def save_to_csv(self):
        with open("cancellation_details.csv", mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([" | ".join(self.names), self.pnr_no, self.journey_date, self.ticket_amount, self.reason, self.refund_mode])
        print("\nCancellation data saved successfully to 'cancellation_details.csv'!")

    def save_to_pdf(self):
        pdf = FPDF()
        pdf.add_page()
        pdf.set_font("Helvetica", "B", 16)
        pdf.cell(0, 10, "Railway Ticket Cancellation Form", new_x="LMARGIN", new_y="NEXT", align='C' )
        pdf.ln(10)
        pdf.set_font("Helvetica", size=12)

        form_text = self.final_output.replace("[bold]", "").replace("[/bold]", "")
        lines = form_text.split("\n")

        for line in lines:
            if line.strip():
                pdf.multi_cell(0, 10, line.strip())
            else:
                pdf.ln(5)

        pdf.output("cancellation_form.pdf")
        print("\nCancellation form PDF generated successfully as 'cancellation_form.pdf'!")



choice = input("Enter mode (Registration / Cancellation): ").lower()
if choice == "registration":
    ob = Registration()
    a = ob.show()
    ob.save_to_csv()
    ob.save_to_pdf()
elif choice == "cancellation":
    ob = Cancellation()
    a = ob.show()
    ob.save_to_csv()
    ob.save_to_pdf()
else:
    print("Invalid choice! Please enter Registration or Cancellation.")

console = Console()
console.print(Panel(a))

```
